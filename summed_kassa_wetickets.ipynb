```python
import pandas as pd
import os
path_1 = os.getcwd()
relative_path = r"WeTickets & Kassa\WeTicketsFinal (version 1)1.xlsx"


path = os.path.join(path_1, relative_path)
df_wetickets = pd.read_excel(path, sheet_name="WeTicketSummed3")


```

```python
df_kassa = df_kassa.rename(columns={'Weeknummer ': 'Weeknummer'})
```

```python
# Functie om tickets te verdelen
def distribute_tickets(relevant_days, extra_tickets):
    indices = relevant_days.index.tolist()
    num_days = len(indices)
    
    if num_days == 0:
        return  # Geen relevante dagen om tickets te verdelen
    
    i = 0
    while extra_tickets != 0:
        if extra_tickets > 0:
            df_wetickets_2.at[indices[i], 'Verkochte_Kaarten'] += 1
            extra_tickets -= 1
        else:
            df_wetickets_2.at[indices[i], 'Verkochte_Kaarten'] -= 1
            extra_tickets += 1
        i = (i + 1) % num_days

# Filteren van de activiteiten van interesse
activiteiten = ["Klimaat/Programmeren", "Workshop Minecraft VR"]

for activiteit in activiteiten:
    df_kassa_filtered = df_kassa[df_kassa['Activiteit'] == activiteit]

    for index, row in df_kassa_filtered.iterrows():
        jaar = row['Jaar']
        weeknummer = row['Weeknummer']
        extra_tickets = row['Tickets_Sold']

        # Vind de relevante rijen in df_wetickets_2
        mask = (df_wetickets_2['Jaar'] == jaar) & (df_wetickets_2['Weeknummer'] == weeknummer) & (df_wetickets_2['Activiteit'] == activiteit)
        relevant_days = df_wetickets_2[mask]

        # Verdeel de tickets over de dagen
        distribute_tickets(relevant_days, extra_tickets)

# Verwijder de rijen uit df_kassa
df_kassa = df_kassa[~df_kassa['Activiteit'].isin(activiteiten)]
```

```python
df_wetickets_2['Bezoekdatum'] = pd.to_datetime(df_wetickets_2['Bezoekdatum'], dayfirst=True).dt.strftime('%d-%m-%Y')
```
