```python
import pandas as pd
import os

path_1 = os.getcwd()
relative_path = r"WeTickets & Kassa\WeTicket tm 16072024.xlsx"
path = os.path.join(path_1, relative_path)



df = pd.read_excel(path, sheet_name="Wet_MC_Filtered_2")
```

```python
df
```

```python
combine_activities = ['Workshop Minecraft programmeren', 'Minecraft workshop Klimaat']
df['Combined Activity'] = df.apply(
    lambda row: 'Klimaat/Programmeren' if row['Activity'] in combine_activities else row['Activity'],
    axis=1)
```

```python
df = df[['Combined Activity'] + [col for col in df.columns if col != 'Combined Activity']]
```

```python
df
```

```python
df = df.groupby(["Combined Activity","Activity","Product","Visit date","Ticket Price"]).agg({'Revenue': 'sum', 'Tickets_Sold': 'sum'}).reset_index()
```

```python
df['Visit date'] = pd.to_datetime(df['Visit date'], dayfirst=True).dt.strftime('%d-%m-%Y')

```

```python
combine_activities = ['Workshop Minecraft programmeren', 'Minecraft workshop Klimaat']


# Separate the dataframe based on the type of activity
df_klimaat = df[(df['Combined Activity'] == 'Klimaat/Programmeren') & (df['Activity'] == 'Minecraft workshop Klimaat')]
df_programmeren = df[(df['Combined Activity'] == 'Klimaat/Programmeren') & (df['Activity'] == 'Workshop Minecraft programmeren')]

# For each row in the df_programmeren, add the values to the matching date row in df_klimaat
for idx, row in df_programmeren.iterrows():
    match_idx = df_klimaat[df_klimaat['Visit date'] == row['Visit date']].index
    if not match_idx.empty:
        df_klimaat.loc[match_idx, 'Revenue'] += row['Revenue']
        df_klimaat.loc[match_idx, 'Tickets_Sold'] += row['Tickets_Sold']
        df_programmeren.drop(idx, inplace=True)

# Combine the modified df_klimaat with the rest of the dataframe
final_df = pd.concat([df_klimaat, df_programmeren])
remaining_df = df[(df['Combined Activity'] != 'Klimaat/Programmeren') | 
                  (df['Activity'] != 'Minecraft workshop Klimaat') & 
                  (df['Activity'] != 'Workshop Minecraft programmeren')]
final_df = pd.concat([final_df, remaining_df], ignore_index=True)
```

```python
final_df['Visit date'] = pd.to_datetime(final_df['Visit date'], dayfirst=True).dt.strftime('%d-%m-%Y')
```

```python
final_df.to_excel("WeTicketsFinal_check.xlsx")
```
